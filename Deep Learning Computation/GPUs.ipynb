Importing Libraries

```
import mxnet as mx
from mxnet import nd
from mxnet.gluon import nn
```

Computing Devices

```
mx.cpu(), mx.gpu(), mx.gpu(1)
```

ND Array and GPUs

```
x = nd.array([1, 2, 3])
x
```

```
x.context
```

Storage on the GPU

```
x = nd.ones((2, 3), ctx=mx.gpu())
x
```

```
y = nd.random.uniform(shape=(2, 3))
y
# we don't have another GPU so consider x in GPU and y in CPU
```

Copying

```
# x + y does not work as there were in different context
z = y.copyto(mx.gpu())
x, z
```

```
x + z
```

```
z = y.as_in_context(mx.gpu())
z
```

```
y.as_in_context(mx.cpu()) is y
```

```
x.copyto(mx.gpu()) is x
```

Gluon and GPUs

```
net = nn.Sequential()
net.add(nn.Dense(1))
net.initialize(ctx=mx.gpu())
```

```
net(x)
```

```
net[0].weight.data()
```
